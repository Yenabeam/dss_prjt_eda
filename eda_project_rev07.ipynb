```python
# 계절 컬럼 추가 
def season(month): 
    if month in [12, 1, 2]:
        return "winter"
    elif month in [3, 4, 5]:
        return "spring"
    elif month in [6, 7, 8]:
        return "summer"
    elif month in [9, 10, 11]:
        return "fall"

df_2016["season"] = df_2016.month.apply(season)
```

```python
# iqr 구하기
def iqr(arr):
    iqr = (np.percentile(arr, 75) - np.percentile(arr, 25)).round(1)
    upper_fence = (np.percentile(arr, 75) + iqr*1.5).round(1)
    lower_fence = (np.percentile(arr, 25) - iqr*1.5).round(1)
    return "iqr: {}, upper_fence: {}, lower_fence: {}".format(iqr, upper_fence, lower_fence)

iqr(date_order["족발/보쌈"]), iqr(date_order["중식"]), iqr(date_order["치킨"]), iqr(date_order["피자"])
```

## 날씨에 따른 주문량 분석

```python
weather = pd.read_csv("./csvfiles/seoul_weather.csv")
weather.rename(columns={"강수량(mm)" : "강수량"}, inplace=True)
weather.head()
```

```python
weather.info()
```

```python
weather = weather.fillna(0)
weather["일시"] = weather["일시"].str.replace("-", "")
weather.rename(columns={"일시" : "일자"}, inplace=True)
weather["일자"] = weather["일자"].astype("int")
weather.head()
```

```python
data = pd.merge(data, weather, how="inner")
```

```python
def weather_func(i):
    if i >= 20:
        return "비"
    elif i >= 10:
        return "흐림"
    else:
        return "맑음"
        
data["날씨"] = data["강수량"].apply(weather_func)
data.head()
```

```python
weather_order = data.groupby(by=["일자", "업종", "날씨"]).mean().reset_index()
weather_order.head()
```
